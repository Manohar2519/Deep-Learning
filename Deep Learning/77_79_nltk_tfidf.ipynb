```python
#Remove special characters
#remove stopwords and perform word_tokenize
from string import punctuation
from nltk.corpus import stopwords
from nltk import word_tokenize
stop_words = stopwords.words("english")
```

```python
#We can add custom stop words
#how to add custom stop words, since my stopwords is a list object I can add items into 
#the list
#We are also adding special characters in stop_words list by converting the punctiation to a list
#and adding it
custom_stop_words = ['jan','feb','mar']

stop_words = stop_words + custom_stop_words+list(punctuation)
print(len(stop_words))
```

#### First perform the tokenization
- Take text as input
- Word tokenize because we are getting the text as string we will do a tokenize
- Convert to lower case
- Remove numbers
- Remove words if they are present in stop_words




```python
def tokenize_text(text):
    words = word_tokenize(text)
    words = [w.lower() for w in words]
    return [w for w in words if w not in stop_words and not w.isdigit()]
```

```python
print(tokenize_text("Jan  5th this class is NLP class 663 d"))
```

```python
len(reuters.fileids())
```

```python
#Yesterday we spoke about 
#Corpus - your entire set of documents
#sample - 1 document
#Vocabulary - all unique words in your corpus is called vocabulary
#Set will only hold unique values
vocab = set()
for file_id in reuters.fileids():
    words = tokenize_text(reuters.raw(file_id))
    vocab.update(words)
vocab = list(vocab)
print(len(vocab))
    
```

```python
#I am creating a word_index
word_index = {w: index for index , w in enumerate(vocab)}
word_index
```

```python
vocab_size = len(vocab)
document_count = len(reuters.fileids())
print(vocab_size,document_count)
```

```python
word_doc_count = np.zeros(vocab_size)
for file in reuters.fileids():
    words = set(tokenize_text(reuters.raw(file)))
    indexes=[word_index[word] for word in words]
    word_doc_count[indexes]+=1.0
```

```python
word_doc_count.argmax()
```

```python
#IDF
#IDF = log(Total_Doc_Count/how_many_docs the word is present)
word_idf = np.log(document_count/(1+word_doc_count).astype(float))
```
