```python
from IPython.display import display_html
import pandas as pd
import numpy as np
def display_side_by_side(*args):
    html_str='  '
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
df1 = pd.DataFrame(np.array((141,158,60,240,21,98,102,195,60)).reshape((3,3)),\
columns=['A','B','C'])

df2 = pd.DataFrame(np.array((158,158,158,240,240,98,195,195,195)).reshape((3,3)),\
columns=['A','B','C'])

display_side_by_side(df1,df2)
```

#### Notes:

 - We will apply a 3 X 1 matrix ,so for computing the first row we will compare between None, 141 and 158 the max is 158 and we will follow the same thing.
 
 - for Dilute process we will take the Max value
 
 - for erode process we will take the min value
#### this is for dilute 
 - Step1a : Max(None,141, 158 ) = 158
 - Step1b : Max(141,158,60) = 158
 - Step1c : Max(158,60,None) = 158
 
#### For erode process
 - Step1a : Min(None,141, 158 ) = 141
 - Step1b : Min(141,158,60) = 60
 - Step1c : Min(158,60,None) = 60

```python
from pylab import *
import cv2
from skimage import io
image = cv2.imread("./data/panther.png")

s1 = 0
s2 = 10
s3 = 15 

t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE

tmp1 = cv2.getStructuringElement(t1,(2*s1+1,2*s1+1),(s1,s1))
tmp2 = cv2.getStructuringElement(t2,(2*s2+1,2*s2+1),(s2,s2))
tmp3 = cv2.getStructuringElement(t3,(2*s3+1,2*s3+1),(s3,s3))

img1 = cv2.dilate(image,tmp1)
img2 = cv2.dilate(image, tmp2)
img3 = cv2.dilate(image, tmp3)

figure(0)
io.imshow(img1)
figure(1)
io.imshow(img2)
figure(2)
io.imshow(img3)

```

```python
#Erode
from pylab import *
import cv2
from skimage import io
image = cv2.imread("./data/panther.png")

s1 = 0
s2 = 10
s3 = 15 

t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE

tmp1 = cv2.getStructuringElement(t1,(2*s1+1,2*s1+1),(s1,s1))
tmp2 = cv2.getStructuringElement(t2,(2*s2+1,2*s2+1),(s2,s2))
tmp3 = cv2.getStructuringElement(t3,(2*s3+1,2*s3+1),(s3,s3))

img1 = cv2.erode(image,tmp1)
img2 = cv2.erode(image, tmp2)
img3 = cv2.erode(image, tmp3)

figure(0)
io.imshow(img1)
figure(1)
io.imshow(img2)
figure(2)
io.imshow(img3)
```
