```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
spam_data = pd.read_csv("spam (2).csv",encoding='latin-1')
```

```python
spam_data.head()
```

```python
spam_data.columns
```

```python
spam_data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1,inplace=True)
spam_data = spam_data.rename(columns={'v1':'target','v2':'text'})
spam_data.head()
```

```python
#You can say this one as text classification 
#On the same lines you will do sentiment analysis?

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(spam_data['text'],spam_data['target'],
                                                test_size=0.2,random_state=1)
```

```python
X_train
```

```python
#CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
vect.fit(X_train)
print("length of vocab",len(vect.get_feature_names()))
```

```python
vect.vocabulary_
```

```python
X_train_vectorized = vect.transform(X_train)
```

```python
# X_train_vectorized.shape
# X_train.shape
# dense=X_train_vectorized.todense()
```

```python
from sklearn.naive_bayes import MultinomialNB
model= MultinomialNB(alpha=0.1)
model.fit(X_train_vectorized,y_train)
```

```python
#Predictions
#Before doing predictions I need to first convert my X_test to vectorized format
X_test_vectorized = vect.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score

train_pred = model.predict(X_train_vectorized)
test_pred = model.predict(X_test_vectorized)
print("Training Accuracy",accuracy_score(y_train,train_pred))
print("Testing Accuracy",accuracy_score(y_test,test_pred))
```

```python
#Lets do it using TFIDF
#min_df minimum document frequency
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(min_df=3).fit(X_train)
print("length of vocab",len(vect.get_feature_names()))
print("longest word",max(vect.vocabulary_,key=len))
```

```python
X_train_vector = vect.transform(X_train)
X_test_vector = vect.transform(X_test)
model= MultinomialNB(alpha=0.1)
model.fit(X_train_vector,y_train)
```

```python
dense_sample = X_train_vector.todense()
dense_sample.shape
```

```python
from sklearn.metrics import accuracy_score

train_pred = model.predict(X_train_vector)
test_pred = model.predict(X_test_vector)
print("Training Accuracy",accuracy_score(y_train,train_pred))
print("Testing Accuracy",accuracy_score(y_test,test_pred))
```
